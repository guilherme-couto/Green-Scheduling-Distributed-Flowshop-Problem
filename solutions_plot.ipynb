```python
import csv
import matplotlib.pyplot as plt

x_array_tft = []
y_array_tec = []
with open('ex.csv', newline='') as csv_file:
    reader = csv.reader(csv_file, delimiter=',', quotechar = '\'')
    next(reader, None)
    for row in reader:
        print(row)
        x_array_tft.append(int(row[1]))
        y_array_tec.append(int(row[2]))

fig,ax=plt.subplots(1,1)
ax.plot(x_array_tft, y_array_tec, 'bo',markersize=6)
ax.set_title('Soluções')
ax.set_xlabel('TFT')
ax.set_ylabel('TEC')
plt.show()
```
