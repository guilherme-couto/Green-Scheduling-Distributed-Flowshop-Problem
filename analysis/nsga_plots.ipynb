```python
import csv
import matplotlib.pyplot as plt
from helpers import get_points_from_file
```

```python
x_tft_initial, y_tec_initial = get_points_from_file('initial_pop_1.csv')
x_tft_after, y_tec_after = get_points_from_file('after_nsga2_1.csv')

fig, ax = plt.subplots(1, 1)
ax.plot(x_tft_initial, y_tec_initial, 'bo', markersize=6, label="initial pop")
ax.plot(x_tft_after, y_tec_after, 'ro', markersize=6, label="after_nsga2")
ax.legend()
ax.set_title('Soluções')
ax.set_xlabel('TFT')
ax.set_ylabel('TEC')
plt.show()
```
